```python
sectors_excerpts = {}
subsectors_excerpts = {}

for i, x in df.iterrows():
    if x['twodim_j'] != 'null':
        objs = json.loads(x['twodim_j'])
        for obj in objs:
            sector = obj['sector'].lower()
            sectors_excerpts[sector] = sectors_excerpts.get(sector, []) + [x['excerpt']]
            if obj['subsectors']:
                for sub in obj['subsectors']:
                    subsectors_excerpts[sub.lower()] = subsectors_excerpts.get(sub.lower(), []) + [x['excerpt']]
```
